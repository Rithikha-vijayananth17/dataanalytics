```python
x1 = np.linspace(-np.pi, np.pi, 256)
x2 = np.linspace(-np.pi, np.pi, 256)
y1 = np.cos(x1)
y2 = np.sin(x2)
```

```python
plt.plot(x1,y1,x2,y2)
plt.show()
```

```python
plt.figure(figsize =(10,20),dpi=100)
plt.plot(x1, y1)
plt.plot(x2, y2)
plt.show()
```

```python
plt.figure(figsize=(12,10),dpi=100)
plt.plot(x1,y1, 'p-.')
plt.plot(x2,y2, 'o--')
plt.show()
```

```python
plt.figure(figsize = (10, 6), dpi = 100)
plt.plot(x1, y1, color = 'purple', linewidth=5)
plt.plot(x2, y2, color = 'orange', linewidth=4)
plt.show()
```

```python
plt.figure(figsize = (10, 6), dpi = 80)
plt.plot(x1, y1, color = 'purple', lw = 2.5, ls = '--')
plt.plot(x2, y2, color = 'green', lw = 2.5, ls = '-.')
plt.show()
```

```python
plt.figure(figsize = (10, 6), dpi = 80)
plt.plot(x1, y1, color = 'purple', lw = 2.5, ls = '--')
plt.plot(x2, y2, color = 'green', lw = 2.5, ls = '-.')
plt.xticks([-np.pi, -np.pi/2, 0, np.pi/2, np.pi])
plt.yticks([-1, 0, 1])
plt.show()
```

```python
plt.figure(figsize = (10, 6), dpi = 80)
plt.plot(x1, y1, color = 'purple', lw = 2.5, ls = '--')
plt.plot(x2, y2, color = 'green', lw = 2.5, ls = '-.')
plt.xticks([-np.pi, -np.pi/2, 0, np.pi/2, np.pi], [r'$-\pi$', r'$-\pi/2$', r'$0$',r'$\pi/2$',r'$\pi$'])
plt.yticks([-1, 0, 1])
plt.show()
```

```python
plt.figure(figsize = (10, 6), dpi = 80)
plt.plot(x1, y1, color = 'purple', lw = 2.5, ls = '--', label = '$y = cosx$')
plt.plot(x2, y2, color = 'green', lw = 2.5, ls = '-.', label = '$y = sinx$')
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Graph')
plt.xticks([-np.pi, -np.pi/2, 0, np.pi/2, np.pi], [r'$-\pi$', r'$-\pi/2$', r'$0$',r'$\pi/2$',r'$\pi$'])
plt.yticks([-1, 0, 1])
plt.legend(loc = 0)
plt.show()
```
