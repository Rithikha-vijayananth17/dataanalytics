```python
import matplotlib.pyplot as plt
import numpy as np
x = np.array([1, 2, 3, 4, 5])
y = np.array([1, 5, 7, 4, 2])
plt.plot(x, y)
plt.xlabel("X-axis Label")
plt.ylabel("Y-axis Label")
plt.title("Simple Line Plot Example")
plt.show()
```

```python
import matplotlib.pyplot as plt
labels = ['Apple', 'Banana', 'Mango', 'Orange']
sizes = [30, 25, 25, 20]
plt.pie(sizes, labels=labels)
plt.title("Fruit Distribution")
plt.show()
```

```python
import matplotlib.pyplot as plt
x = [1, 2, 3, 4, 5]
y = [5, 3, 7, 1, 9]
plt.scatter(x, y)
plt.title("Scatter Plot")
plt.xlabel("X")
plt.ylabel("Y")
plt.show()
```

```python


import matplotlib.pyplot as plt
x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]
z = [1, 3, 2, 5, 4]
plt.plot(x, label='x')
plt.plot(y, label='y')
plt.plot(z, label='z')
plt.title("x, y, z on Same Plot")
plt.xlabel("Index")
plt.ylabel("Values")
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]
z = [1, 3, 9, 16, 25]
plt.plot(x, y, color='red', linestyle='--', linewidth=2)
plt.plot(x, z, color='blue', linestyle=':', linewidth=3)
plt.title("Styled Line Plot")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.legend()

plt.show()
```
